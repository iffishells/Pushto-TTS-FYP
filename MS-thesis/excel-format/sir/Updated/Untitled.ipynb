```python
## importing lib's

import nltk
import re
import pandas as pd
import termcolor as colored
```

```python
# importing data or reading data sets files

# read pashto dict 
dic = pd.read_csv('dictionary.csv')
# read corpus data for tagging the sentance unabelled
doc = pd.read_csv('corpus.csv' ,quotechar='"' , delimiter=',')

doc.columns = ["original" , "Tagged"] # two more colum in doc files

def grammar_to_tag(arg):
    '''
    grammer to tag function assign the short form to each tag
    
    '''
    switcher={
            "Noun": "N",
            "Verb": "V",
            "Adverb": "ADV",
            "Adjective": "ADJ",
            "Numerical": "NUM",
            "ProNoun" : "PRN",
            "Abbreviation": "N",
            "Interjection" : "INT",
            "Particle" : "RP",
            "Preposition" : "PrP",
            "Postposition" : "PoP",
            "Suffix" : "SUF",
            "Prefix" : "PRF",
            "Conjunction" : "CO"
            }
    return switcher.get(arg,'none')

def number_to_pos(arg):
    
    '''
        Number_to_pos is a function to assign the number to each tags
    '''
    switcher={
            1: "Noun",
            2: "Verb",
            3: "Adverb",
            4: "Adjective",
            5: "Numerical",
            6: "ProNoun",
            7: "Abbreviation",
            8: "Interjection",
            9: "Particle",
            10: "Preposition",
            11: "Postposition",
            12: "Suffix",
            13: "Prefix",
            14: "Conjunction"
            }
    return switcher.get(arg,"none")

# sentace tuple
sentance_tuple = []

# sentace string
sentance_string =""

# control counter 
c =0
for index , line_csv in doc.iterrows():
    
    #line 
    line = line_csv[0]
    print("Line : ",line)
    #pos
    pos = line_csv[1]
    
    print("POS : ",pos)
    if c>5:
        break

    if isinstance(pos , str):
        print("Pos : ",pos) # printing without pre-processing
        
        tags = [nltk.tag.str2tuple(t) for t in pos.split()]
        
        print("Tag : (with pre-prcessing) " ,tags )
        continue
        
    line = re.sub("\,", "", line) # remove commas
    line = re.sub("\،", "", line) # remove commas
    line = re.sub("\‘", "", line) # remove weird symbol
    line = re.sub("\’", "", line) # remove weird symbol
    line = re.sub("\۔", "", line) # remove weird symbol
    
    sentences = re.split(r' *[\.\?!][\'"\)\]]* *', line)
    
    for sentence in sentences:
        
        if len(sentence) >0 :
#             print(colored(sentence,'red'))
            print("Len > 0")
            
            
        words = nltk.tokenize.word_tokenize(sentence)
        words_in_sentance = []
        
        full_version =0
        
        for w in words:
            pos = dic['WORD'] == w
            print("--> ",pos)
            if len(dic[pos]) ==1:
                
                w1 = dic['WORD'][pos].to_string(header = False,index = False)
                w2 = dic["POS"][pos].to_string(header=False,index=False)
                w3 = grammar_to_tag(w2)
                print("w1 , w2 , w3 ",w1,w2,w3)
                
                words_in_sentance.append([w1.strip(),w2.strip(),w3.strip()])

            if len(dic[pos] >1):
                w1 = dic["WORD"][pos].to_string(header=False,index=False)
                w2 = dic["POS"][pos].to_string(header=False,index=False)
                temp = []
                
                count = 1
                
                for (f,b) in zip(w1.split() ,w2.split()):
                    
                    w3 = grammar_to_tag(b)
                    print("%2s:%15s\t(%s) %s"% (count, f, w3, b))
                    temp.append([f,b,w3])
                    count +=1
                print("%2s:%15s\t%s"% (count, "نور", "other"))
                
                inp = int(input("Which POS for ("+colored(w,'red')+") is correct?"))
                
                if inp ==count:
                    full_version =1
                    
                else:
                    words_in_sentance.append([temp[inp-1][0].strip(),temp[inp-1][1].strip(),temp[inp-1][2].strip()])

                    
                    
                
                
                
    
    
```
